```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import imageio

from scipy import ndimage
```

```python
image = imageio.v2.imread('me.JPG')

image = ndimage.rotate(image, -90, reshape=True)

photo_data = np.array(image)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(photo_data)
```

```python
photo_data[1400:1600, 650:1500] = 0
plt.figure(figsize=(15,15))
plt.imshow(photo_data)
```

```python
imageio.imwrite("censored_me.jpg", photo_data)
```
