### Pandas series

Pandas *Series* is one-dimentional labeled array containing data of the same type (integers, strings, floating point numbers, Python objects, etc. ). The axis labels are often referred to as *index*.

```python
# Example of creating Pandas series :
s1 = pd.Series( np.random.randn(5), dtype = 'float32')
print(s1)
```

We did not pass any index, so by default, it assigned the indexes ranging from 0 to len(data)-1

```python
# View index values
print(s1.index)
print(type(s1.index))

#How to obtain s1.index in list?
```

```python
s2 = pd.Series([1,2,3,4,'5'], index=['a', 'b', 'c', 'd', 'e'])
print(s2['a'])
print(s2.dtypes)
print(s2.index)
print(type(s2.index))
print(list(s2.index))
```

```python
# Creating Pandas series with index:
s2 = pd.Series( np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'] )
print(s2)
```

```python
# View index values
print(s2.index)
print(type(s2.index))
```

```python
# Create a Series from dictionary
data = {'pi': 3.1415, 'e': 2.71828}  # dictionary
print(data, type(data))
print("-"*30)
s3 = pd.Series ( data )
print(s3, type(s3))
```

```python
#What will happen? Reindexed
s4 = pd.Series ( data, index = ['e', 'pi', 'tau'])
print(s4)
```

NAN (non a number) - is used to specify a missing value in Pandas.

```python
# Creating a Pandas Series object from a single number:
s5 = pd.Series( 1, index = range(10), name='Ones')
print(s5)
print("-"*30)
print(s5.name)
s5.name = "Ini"
print(s5.name)
#Series has a name attribute
```

The Series name can be assigned automatically in many cases, in particular, when selecting a single column from a DataFrame, the name will be assigned the column label.

```python
# Many ways to "slice" Pandas series (series have zero-based index by default):
s1 = pd.Series(np.random.randn(5))
print(s1)
s1[3]  # returns 4th element
```

```python
s1[:2] # First 2 elements
```

```python
print( s1[ [2,1,0]])  # Elements out of order
print(type(s1[[2,1,0]]))
```

```python
#Slicing series using index label (access series like a dictionary)
s4 = pd.Series(data, index = ['e','pi','tau'])
print(s4)
print(s4['pi'])
print(s4[1])
```

### Pandas DataFrame

Pandas *DataFrame* is two-dimensional, size-mutable, heterogeneous tabular data structure with labeled rows and columns ( axes ). Can be thought of a dictionary-like container to store python Series objects.

DataFrame structure
structure_table.jpg

```python
import numpy as np
import pandas as pd
matrix_data = np.random.randint(1,20,size=20).reshape(5,4)
row_labels = ['A', 'B', 'C','D','E']
column_headings = ['W','X','Y','Z']
df2 = pd.DataFrame(data = matrix_data, index = row_labels, \
                   columns=column_headings)
df2
```

```python
df2[df2['W']==12].X
```

```python
matrix_data = np.random.randint(1,20,size=20).reshape(5,4)
column_headings = ['W','X','Y','Z']
df2 = pd.DataFrame(data = matrix_data, columns=column_headings)
df2
```

```python
df2.loc[1:3]
```

```python
df2.iloc[1:3]
```
