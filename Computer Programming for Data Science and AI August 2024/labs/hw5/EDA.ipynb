#### Salary vs Location | Get average salary for all the location and write out the top 3 and botton 3 location according to salaries (5 marks)

```python
saloc = salary_data.groupby('company_location').salary_in_usd.mean().reset_index().sort_values(by='salary_in_usd', ascending=False)
saloc.head(3), saloc.tail(3)
```
