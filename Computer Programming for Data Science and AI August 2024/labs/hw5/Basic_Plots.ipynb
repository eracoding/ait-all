The explicit and the implicit interfaces

```python
#The explicit interface
x = np.linspace(0, 2, 100)  # Sample data.

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
#The implicit interface
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

### Creating Figures
You can use __plt.figure()__ to create a new Figure. This function returns a Figure instance, but it is also passed to the backend.
<br><br>
By default, the Figure has a width of 6.4 inches and a height of 4.8 inches with a dpi (dots per inch) of 100. To change the default values of the Figure, we can use the parameters figsize and dpi.
<br><br>
Even though it is not necessary to explicitly create a Figure, this is a good practice if you want to create multiple Figures at the same time.

```python
#To change the width and the height
plt.figure()
plt.figure(num = 2, figsize=(8, 5), dpi = 300)
```

### Closing Figures
Figures that are no longer used should be closed by explicitly calling __plt.close()__, which also cleans up memory efficiently.
<br>
The plt.close() command will close the current Figure. To close a specific Figure, you can either provide a reference to a Figure instance or provide the Figure number.

```python
#Create Figure with Figure number 10
plt.figure(num=10)
#Close Figure with Figure number 10
plt.close(10)
```

The plt.close('all') command is used to close all active Figures.

```python
plt.close('all')
```

### Displaying SubFigures

Sometimes we want to have a nested layout in a Figure, with two or more sets of Axes that do not share the same subplot grid. We can use add_subfigure or subfigures to create virtual figures inside a parent Figure

```python
fig = plt.figure(layout='constrained', facecolor='lightskyblue')
fig.suptitle('Figure')
figL, figR = fig.subfigures(1, 2)
figL.set_facecolor('thistle')
axL = figL.subplots(2, 1, sharex=True)
axL[1].set_xlabel('x [m]')
figL.suptitle('Left subfigure')
figR.set_facecolor('paleturquoise')
axR = figR.subplots(1, 2, sharey=True)
axR[0].set_title('Axes 1')
figR.suptitle('Right subfigure')
```

## Layout
### Subplots

```python
import numpy as np
series = []
for i in range(4):
    series.append(np.random.randint(low = -5+i*2, high=10, size = 50))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
#returns contiguous flattened array (1D array)

axes = axes.ravel() #flatten
for i, ax in enumerate(axes): #add a counter i to iterator axes
    ax.plot(series[i])
    ax.set_title('Subplot ' + str(i))
```

```python
#Setting sharex and sharey to True results in the following diagram. This allows for a better comparison:
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
axes = axes.ravel()
for i, ax in enumerate(axes):
    ax.plot(series[i])
```

```python
# show the same result as previous cell
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.plot(series[i])
    plt.title('Subplot ' + str(i))
plt.show()
plt.close()
```

### Tight Layout

The plt.tight_layout() adjusts subplot parameters (primarily padding between the Figure edge and the edges of subplots, and padding between the edges of adjacent subplots) so that the subplots fit well in the Figure.

```python
# No tight_layer()
fig, axes = plt.subplots(2, 2)
axes = axes.ravel()
for i, ax in enumerate(axes):
    ax.plot(series[i])
    ax.set_title('Subplot ' + str(i))
```

```python
# call tight_layer()
fig, axes = plt.subplots(2, 2)
axes = axes.ravel()
for i, ax in enumerate(axes):
    ax.plot(series[i])
    ax.set_title('Subplot ' + str(i))
fig.tight_layout()
```

### GridSpec
The __matplotlib.gridspec.GridSpec(nrows, ncols)__ function specifies the geometry of the grid in which a subplot will be placed.

```python
fig = plt.figure()
gs = fig.add_gridspec(3, 4)
ax1 = fig.add_subplot(gs[:3, :3])
ax2 = fig.add_subplot(gs[0, 3])
ax3 = fig.add_subplot(gs[1, 3])
ax4 = fig.add_subplot(gs[2, 3])
ax1.plot(series[0])
ax2.plot(series[1])
ax3.plot(series[2])
ax4.plot(series[3])
fig.tight_layout()
```

```python
gs = mpl.gridspec.GridSpec(3, 4)
ax1 = plt.subplot(gs[:3, :3])
ax2 = plt.subplot(gs[0, 3])
ax3 = plt.subplot(gs[1, 3])
ax4 = plt.subplot(gs[2, 3])
ax1.plot(series[0])
ax2.plot(series[1])
ax3.plot(series[2])
ax4.plot(series[3])
plt.tight_layout()
```

### Basic Text and Legend Functions

#### Labels
Matplotlib provides a few label functions that we can use for setting labels to the x- and y-axes. <br>
The __plt.xlabel()__ and __plt.ylabel()__ functions are used to set the label for the current axes <br>
The __set_xlabel()__ and __set_ylabel()__ functions are used to set the label for specified axes.

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
# the histogram of the data
n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
```

### Bar Chart
<br>
The plt.bar(x, height, [width]) creates a vertical bar plot. For horizontal bars, use the plt.barh() function.
<br>
x : Specifies the x coordinates of the bars<br>
height : Specifies the height of the bars<br>
width (optional): Specifies the width of all bars; the default is 0.8<br>

```python
fig, ax = plt.subplots(figsize=(6,4), layout='constrained')
category = ['A', 'B', 'C', 'D']
ax.bar(category, [20, 25, 40, 10], label="Label")
ax.legend()
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title('Title', fontsize = 14)
```

### Line plot


```python
#The flights dataset has 10 years of monthly airline passenger data
flights = pd.read_csv(f"{path}/flight_details.csv")
flights.head()
```

```python
#To draw a line plot using long-form data, assign the x and y variables:
may_flights = flights.query("Months == 'May'")
sns.lineplot(data=may_flights, x="Years", y="Passengers")
plt.show()
```

### Removing Axes Spines

Sometimes, it might be desirable to remove the top and right axes spines. The despine() function is used to remove the top and right axes spines from the plot

```python
month_filter = flights[flights['Months'].isin(['January','February','March','April','May'])]
sns.set_style("white")
#lp is a subplot
lp = sns.lineplot(data=month_filter, x="Years", y="Passengers", hue='Months')
lp.legend(loc='upper left')

#Remove the axes spines
sns.despine()
plt.show()
plt.close()
```

### Heatmaps
The visualization of multivariate data can be done using heatmaps as they are great for finding patterns in your data.

```python
plt.figure()
sns.heatmap(data, cmap=sns.cubehelix_palette(rot=-.3, as_cmap=True)) # you can use any sequential color palette
plt.title("Flight Passengers from 1949 to 1960")
plt.show()
plt.close()
```

### Box Plot

```python
mydata = pd.read_csv(f"{path}/iq_scores.csv")
mydata.head()
```

Access the data of each test group in the column. Convert them into a list using the tolist() method. Once the data of each test group has been converted into a list, assign this list to variables of each respective test group:

```python
group_a = mydata[mydata.columns[0]].tolist()
group_b = mydata[mydata.columns[1]].tolist()
group_c = mydata[mydata.columns[2]].tolist()
group_d = mydata[mydata.columns[3]].tolist()
```

construct a DataFrame from this given data

```python
data = pd.DataFrame({'Groups': ['Group A'] * len(group_a) + ['Group B'] * len(group_b) + ['Group C'] * len(group_c) + ['Group D'] * len(group_d),
                     'IQ score': group_a + group_b + group_c + group_d})
data.head()
```

```python
#another way to prepare data
mydata_2=pd.melt(mydata, value_vars=mydata.columns, ignore_index=False,
                 var_name='Groups', value_name='IQ score')
mydata_2.head()
```

```python
#plt.figure(dpi=180)
# Set style to 'whitegrid'
sns.set_style('whitegrid')
# Create boxplot
sns.boxplot(x='Groups', y='IQ score', data=data, hue='Groups')
# Despine
sns.despine(left=True, right=True, top=True)
# Add title
plt.title('IQ scores for different test groups')
# Show plot
plt.show()
plt.close()
```

### Violin Plots

Create a violin plot for each of the IQ scores of different test groups using Seaborn's violinplot function. Use the whitegrid style and remove all axes splines, except the one on the bottom. Add a title.

```python
# Create figure
plt.figure()
# Set style
sns.set_style('whitegrid')

# Create violin plot
sns.violinplot(x='Groups', y='IQ score', data=data, hue='Groups')
# Despine
sns.despine(left=True, right=True, top=True)
# Add title
plt.title('IQ scores for different test groups')
# Show plot
plt.show()
plt.close()
```
